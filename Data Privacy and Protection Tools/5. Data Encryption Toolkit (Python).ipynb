```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import base64
import os

# Function to pad data to fit AES block size
def pad_data(data):
    block_size = AES.block_size
    padding_length = block_size - len(data) % block_size
    return data + bytes([padding_length]) * padding_length

# Function to encrypt a file
def encrypt_file(file_path, key):
    cipher = AES.new(key, AES.MODE_CBC)
    
    with open(file_path, 'rb') as file:
        file_data = file.read()
    
    # Pad data to ensure it fits AES block size
    padded_data = pad_data(file_data)
    
    encrypted_data = cipher.encrypt(padded_data)
    
    # Save encrypted data and IV
    with open(f"{file_path}.enc", 'wb') as enc_file:
        enc_file.write(cipher.iv)
        enc_file.write(encrypted_data)
    
    print(f"File {file_path} encrypted successfully.")

# Function to decrypt a file
def decrypt_file(file_path, key):
    with open(file_path, 'rb') as enc_file:
        iv = enc_file.read(16)
        encrypted_data = enc_file.read()

    cipher = AES.new(key, AES.MODE_CBC, iv)
    decrypted_data = cipher.decrypt(encrypted_data)
    
    # Remove padding
    padding_length = decrypted_data[-1]
    decrypted_data = decrypted_data[:-padding_length]
    
    # Save decrypted file
    decrypted_file_path = file_path.replace(".enc", ".dec")
    with open(decrypted_file_path, 'wb') as dec_file:
        dec_file.write(decrypted_data)
    
    print(f"File {file_path} decrypted successfully.")

# Generate a random AES key
key = get_random_bytes(16)

# Example usage
file_path = "sample_data.txt"
encrypt_file(file_path, key)
decrypt_file(f"{file_path}.enc", key)
```
