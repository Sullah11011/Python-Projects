```python
import gym
import numpy as np
import random

class RedTeamEnv(gym.Env):
    def __init__(self):
        super(RedTeamEnv, self).__init__()
        self.action_space = gym.spaces.Discrete(4)  # Four actions: exploit vulnerability, social engineering, phishing, reconnaissance
        self.observation_space = gym.spaces.Discrete(5)  # Five network states: low security, medium security, high security, breached, detected
        self.state = 0  # Initial state: low security

    def reset(self):
        self.state = random.choice([0, 1, 2, 3])  # Random security state
        return self.state

    def step(self, action):
        state = self.state
        reward = 0
        done = False

        if action == 0:  # Exploit vulnerability
            if state == 1:  # Medium security
                reward = 10
                self.state = 3  # Breached
            elif state == 2:  # High security
                reward = -1
                done = True
            else:
                reward = -1

        elif action == 1:  # Social engineering
            if state == 0:  # Low security
                reward = 5
                self.state = 1  # Medium security
            elif state == 3:  # Breached
                reward = 10
                done = True
            else:
                reward = -1

        elif action == 2:  # Phishing
            if state == 1:  # Medium security
                reward = 10
                self.state = 2  # High security
            else:
                reward = -1

        elif action == 3:  # Reconnaissance
            if state != 4:  # If not detected
                reward = 2
            else:
                reward = -5  # If detected during reconnaissance

        return self.state, reward, done, {}

    def render(self):
        state_map = ['Low Security', 'Medium Security', 'High Security', 'Breached', 'Detected']
        print(f"Red Team State: {state_map[self.state]}")
```
