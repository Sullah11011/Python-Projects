```python
from flask import Flask, render_template, request, redirect, url_for
import pandas as pd
import plotly.express as px
import plotly.io as pio

app = Flask(__name__)

# Simple login system (this should be expanded in real-world scenarios)
users = {"admin": "password123"}

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        username = request.form["username"]
        password = request.form["password"]
        if users.get(username) == password:
            return redirect(url_for("dashboard"))
        else:
            return "Login failed! Check your credentials."
    return render_template("login.html")

@app.route("/dashboard")
def dashboard():
    # Example cybersecurity data (you can fetch this from APIs or databases)
    data = {
        "Metric": ["Firewalls", "Intrusions", "Malware", "Patches", "User Logins"],
        "Value": [85, 5, 1, 10, 50],
        "Threshold": [80, 2, 0, 15, 40]
    }
    
    df = pd.DataFrame(data)
    
    # Create a bar chart for the data
    fig = px.bar(df, x="Metric", y="Value", title="Cybersecurity Dashboard Metrics")
    fig.update_layout(showlegend=False)
    
    # Render the plotly figure in HTML
    graph_html = pio.to_html(fig, full_html=False)
    
    return render_template("dashboard.html", graph_html=graph_html)

if __name__ == "__main__":
    app.run(debug=True)
```
